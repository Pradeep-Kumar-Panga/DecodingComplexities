```python
import numpy as np

X_raw = np.array([[1], [2], [4], [5], [6]])
y = np.array([[45], [50], [75], [80], [95]])

print(X_raw)
print(y)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_raw, y)
plt.show()
```

```python
# Add the Bias column (ones)
X = np.c_[np.ones(X_raw.shape[0]), X_raw]
print(X)
```

```python
# X Transpose X and X Transpose y
X_T_X = np.dot(X.T, X)
X_T_y = np.dot(X.T, y)

print(X_T_X)
print(X_T_y)
```

```python
# Calculate the inverse of X Transpose X
X_T_X_inv = np.linalg.inv(X_T_X)

# Multiply X Transpose X Inverse with X Transpose y
w = np.dot(X_T_X_inv, X_T_y)
print(w)
```

```python
# Solve using 'solve' method
w = np.linalg.solve(X_T_X, X_T_y)
print(w)
```

```python
# Predict score if student has not studied
X_pred = np.array([[1, 0]])
y_pred = np.dot(X_pred, w)
print(y_pred)
```

```python
# Predict score for all inputs
y_pred = np.dot(X, w)
print(y_pred)
```

```python
# Draw regression line
plt.scatter(X_raw, y)
plt.plot(X_raw, y_pred, color='red')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Initialize and fit
# Note: sklearn handles the intercept automatically, so we pass X_raw
model = LinearRegression()
model.fit(X_raw, y)

print("--- Comparison ---")
print(f"Our Intercept: {w[0][0]:.4f}")
print(f"Sklearn Intercept: {model.intercept_[0]:.4f}")

print(f"Our Slope: {w[1][0]:.4f}")
print(f"Sklearn Slope: {model.coef_[0][0]:.4f}")
```
